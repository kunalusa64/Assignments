```python
#Q1.1
# reference for code from class notes and recorded videos
# importing libraries
import pandas as pd
import numpy as np
# assigning path 
path = 'https://docs.google.com/spreadsheets/d/e/2PACX-1vSuPjI7TrhV3FhZ8syEtPCHdtlBLbnNzBujxOinZiXsaxeQkMkT0C7LyTgL8Tqx5KyhTv0Vj0j7GnhX/pub?gid=464058113&single=true&output=csv'
a = pd.read_csv(path) # loading matric into a
N=10 # order of matrix
coeff = a.iloc[:,:N] # assigning coefficient of matrix
rhs = a.iloc[:,N:N+1] # assigning rhs of matrix
det = np.linalg.det(coeff) # calculating determinant of matrix
"""
The research paper uses two transformations of which the below code is for first transformation
The ptrans1 transformation converts matrix into a form and uses backward substitution
We have defined a function named ptrans1 to solve the pentadiagonal matrix
then we calculate the determinant of matrix to know whether the matrix is singular or not
We initialize our vectors and then get values for each element of the array using the
equations in the algorithm
"""
# defining a function to solve the pentadiagonal matrix
def ptrans1(n,d,a,b,c,e,y):
    # Performing if condition to know what whether the matrix is singular or not
  if det == 0: 
    print('Matrix is singular')
  else:
    # initializing our vectors as per the research paper
    # creating an array of size n
    # coeffiecient of matrix
      mu = np.zeros(n,dtype = 'float64')
      alpha = np.zeros(n,dtype = 'float64')
      beta = np.zeros(n,dtype = 'float64')
      gama = np.zeros(n,dtype = 'float64')
      # rhs of matrix
      z = np.zeros(n,dtype = 'float64')
      x = np.zeros(n,dtype = 'float64')
      """
      After initializing our vectors, we perform elementary row operation to convert
      the matrix into a form that is easy to solve using backward substitution
      """
      # setting up equations for as per research paper
      # getting values for n = 1st term of array
      mu[0] = d[0]
      alpha[0] = a[0]/mu[0]
      beta[0] = b[0]/mu[0]
      z[0] = y[0]/mu[0]
     
      # getting values for n = 2nd term of array
      gama[1] = c[1]
      mu[1] = d[1] - (alpha[0]*gama[1])
      alpha[1] = (a[1] - (beta[0]*gama[1]))/mu[1]
      beta[1] = b[1]/mu[1]
      z[1] = (y[1] - (z[0]*gama[1]))/mu[1]   
     
     # Lopping through to get values for n from i= 3rd term,.......,(n-2)th term
      for i in np.arange(2,(n-2),1):
         gama[i] = c[i] - (alpha[i-2]*e[i])
         mu[i] = d[i] - (beta[i-2]*e[i]) - (alpha[i-1]*gama[i])
         alpha[i] = (a[i] - (beta[i-1]*gama[i]))/mu[i]
         beta[i] = b[i]/mu[i]
         z[i] = (y[i] - (z[i-2]*e[i]) - (z[i-1]*gama[i]))/mu[i]
     
     # getting values for n = (n-1) term of array
      gama[n-2] = c[n-2] - (alpha[n-4]*e[n-2])
      mu[n-2] = d[n-2] - (beta[n-4]*e[n-2]) - (alpha[n-3]*gama[n-2])
      alpha[n-2] = (a[n-2] - (beta[n-3]*gama[n-2]))/mu[n-2]
      z[n-2] = (y[n-2] - (z[n-4]*e[n-2]) - (z[n-3]*gama[n-2]))/mu[n-2]
     
     #for values for n = nth term of array
      gama[n-1] = c[n-1] - (alpha[n-3]*e[n-1])
      mu[n-1] = d[n-1] - (beta[n-3]*e[n-1]) - (alpha[n-2]*gama[n-1])
      z[n-1] = (y[n-1] - (z[n-3]*e[n-1]) - (z[n-2]*gama[n-1]))/mu[n-1]
     
     # computing vector X
     # performing backward substitution 
      x[n-1] = z[n-1] # getting values of vector X for n = nth term 
      x[n-2] = z[n-2] - (alpha[n-2]*x[n-1]) # getting values of vector X for n = (n-1)th term 
     # lopping through to get values for n from i = (n-2)th term,....1st term of array
      for i in np.arange((n-3),-1,-1):
         x[i] = z[i] - (alpha[i]*x[i+1]) - (beta[i]*x[i+2])  
      return(x) # return all values of X 

# initializing our diagonal elements
n= 10 # order of matrix
# creating an array of zeros of size n
a = np.zeros(n,dtype='float64') 
b = np.zeros(n,dtype='float64')
c = np.zeros(n,dtype='float64')
d = np.zeros(n,dtype='float64')
e = np.zeros(n,dtype='float64')
y = np.zeros(n,dtype='float64')
# values of diagonal elements of given matrix as per research paper
d = np.diag(coeff) # values of longest diagonal of given matrix(central most diagonal)
a = np.diag(coeff, k=1) # diagonal next to main diagonal on upper triangular matrix
b = np.diag(coeff, k=2) # diagonal next to a
c = (0,3,2,1,2,1,2,1,-2,4) # diagonal next to main diagonal on lower triangular matrix
e = (0,0,1,3,1,5,2,2,2,-1) # diagonal next to c
y = (8,33,8,24,29,98,99,17,57,108) # values of the rhs matrix# Calling function 
X = np.round(ptrans1(n,d,a,b,c,e,y),0) 
print('X = ',X) # printing vector X
```

# ***Code for Sor method***

```python
#Q2:
"""
Code reference from class notes and recorded lectures and the below
'https://en.wikipedia.org/wiki/Successive_over-relaxation#Algorithm'
"""
#importing libraries
import numpy as np
import os
import pandas as pd
# defining a function to solve the matrix of research paper using sor method
def sor(coeff, rhs, w, init, converg):
    """
        coeff: 
        rhs: n dimensional numpy vector.
        w: relaxation factor.
        init: An initial solution guess for the solver to start with.
        converg: The maximum discrepancy acceptable to regard the 
        current solution as fitting.
    Returns:
        x: solution vector of dimension n.
    """
    # initialization
    step = 0 
    x = init[:] # initialization of resultant vector
    residual = np.linalg.norm(np.matmul(coeff, x) - rhs)  # getting Initial residual
    while residual > converg:  # until residual > converg run while loop
        for i in range(coeff.shape[0]):
            term1 = 0 #initialize term1
            for j in range(coeff.shape[1]):
                if j != i:
                    term1 += (coeff[i, j] * x[j]) # calculate term1 required to calculate x[i]
            x[i] =  x[i] + (w / coeff[i, i]) * (rhs[i] - term1) # equation for Sor
        residual = np.linalg.norm(np.matmul(coeff, x) - rhs)
        step += 1 # increment
    return (x)

# assigning path 
path = 'https://docs.google.com/spreadsheets/d/e/2PACX-1vSuPjI7TrhV3FhZ8syEtPCHdtlBLbnNzBujxOinZiXsaxeQkMkT0C7LyTgL8Tqx5KyhTv0Vj0j7GnhX/pub?gid=464058113&single=true&output=csv'
a = pd.read_csv(path) # loading matric into a
N=10 # order of matrix

residual_convergence = 1e-10
w = 1.2 # Relaxation factor
coeff = a.iloc[:,:N] # coefficient matrix
rhs = a.iloc[:,N:N+1]# rhs matrix
init = np.zeros(N) # initialization 
x = sor(coeff, rhs, w, init, residual_convergence) # calling function sor
# printing our resultant vector
print(x)
```
