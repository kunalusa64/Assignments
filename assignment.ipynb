```python
Q1.1 – TRAPEZOIDAL EQUATION
#import numpy library
import numpy as np
#defining a function to calculate the value of given equation 
def fx(x):
    z = np.sqrt(1+x**2) # our given equation
    return z # returning the value backk to function
# integration limits
xo = 0 # lower limit
xn = 1 # upper limit
n = 10 # number of sub-intervals
fxa = fx(xo) # substituting the value of xo in given equaton
fxb = fx(xn) # substituting the value of xn in given equaton
h = (xn-xo)/n # calculating the size of steps
# Finding the sum of fx after substituting the values into our equation
sum = fxa + fxb 
# running a for loop to calculate the value of f(k)
for i in range(1,n):
    k = xo + i*h
    sum = sum + 2 * fx(k) # finding the summation of the term of composite trapezoidal equation
# calculating our final value of composite trapezoidal integration
inti = 0.5*h*sum   
print(inti) # printing our result






Q1.2 – SIMPSON’S  EQUATION
#import numoy library
import numpy as np
#defining a function to calculate the value of given equation 
def fx(x):
    z = np.sqrt(1+x**2) # our given equation
    return z # returning the value backk to function
# sum limits
xo = 0 # lower limit
xn = 1 # upper limit
n = 20 # number of sub-intervals
fxa = fx(xo) # substituting the value of xo in given equaton
fxb = fx(xn) # substituting the value of xn in given equaton
h = (xn-xo)/n # calculating the size of steps
# Finding the sum of fx after substituting the values into our equation
sum = fxa + fxb 
# running a for loop to calculate the value of f(k)
for i in range(1,n):
    k = xo + i*h
#if statements to calculate the values of 2nd and 3nd term in the composite equation
    if i%2 == 0: #if the remainder is even calculate the following sum
            sum = sum + 2 * fx(k) # 
    else: #if the remainder is odd calculate the following sum
            sum = sum + 3 * fx(k)
# calculating our final value of simpson's composite equation
inti = 3*h*sum/8   
print(inti) # printing our result




Q1.3 – Romberg’S  EQUATION
"""
Getting assistance for romberg integration method from the lecture notes and code
provided to calculate the intergration of the given equation
"""
#importing libraries
import numpy as np
"""
We define a function fx whose integration is to be found out
"""
# Function to compute integrand
def fx(x):
    z = np.sqrt(x**2+1) # our given equation
    return(z)
""" 
We provide the upper and lower limits and the order(N) upto which the function
needs to calculate the integrand
"""
#define order and limits of integration
N = 10 # order of romberg integral
b = 1.0 # upper limit
a = 0.0 # Lower limit
""" 
We create a matrix of size of the order(N) to store values of the all integrand 
elements we would be calculating further
and we substitute teh values of upper and lower limits into our given equation
and then calculate the sum of those values
"""
# Define matrix to store values
T = np.zeros([N,N]) # initialize results matrix of order 10x10
fxa = fx(a) # compute the integrand at LL
fxb = fx(b) # compute the integrand at UL
sum1 = fxa + fxb # Sum the values obtained after substituting UL and LL values
"""
We calculate the width of initial panel for romberg
then we calculate the value of first element of first column which will be 
used further to calculate remaning values
and we create an array to run a for loop 
"""
h = b-a # intial panel width
T[0,0] = h/2*(sum1) # compute single panel estimate
K = np.arange(1,N,1) # create an array of k
""" 
We run a for loop to calculate terms of first column using composite trapezoidal 
and then extrapolate the values using values from previous column.
We divide our general T(j,k) into parts, calculate them separately and then
calculate the sum.
"""
for k in K:
    h = (b-a)/2**k #update the panel width
    idx = np.arange(1,(2**k)) #create summation idx
    z = a + idx*h # create intermediate values
    fsum = np.sum(fx(z)) # sum intermediate values
    T[k,0] = h/2*(sum1 + 2*fsum ) # apply trapezoidal rule
    for j in np.arange(1,k+1,1): # Apply Richardson extrapolation
       num1 = 1/(4**j - 1)
       num2 = 4**j*T[(k),(j-1)]
       num3 = T[(k-1),(j-1)]
       T[k,j] = num1*(num2-num3) # obtain refined estimates
""" 
In romberg integration we know that the diagonal elements has the most accurate 
values and hence we filter out the diagonal values from T
and then we loop through to print our claculated integral values till our 
specified order(N)
"""
Res = np.diagonal(T) # collect diagonal terms
# Loop through and print diagonal values
idx = 0
for m in Res:
   idx = idx + 1
   print('Iter:', idx ,'Result',m)


    
    
Q1.4 Guass Quadrature equation
"""
Getting assistance for Guass quatrature integration method from the lecture 
notes and code provided to calculate the intergration of the given equation
"""
# import libraries
import numpy as np
import pandas as pd
import os
'''
In this code what we do is we define a function for the given equation
then we define a function to implement quass quadrature equation
In this code we have imported the values of weights and abscissa(upto 
the order n=10) that are required to integrate our given equation.
We found those from the wesite below
https://keisan.casio.com/exec/system/1329114617
'''
# Function to be integrated
def fx(x):
   z = np.sqrt(1+x**2)
   return(z)
# Function for coordinate transformation
def newcoord(z,a,b):
   xx = 0.5*((b-a)*z+(a+b))
   return xx
"""
We set the path, change the working directory and read the file we imported. 
We extract the columns of weights(Xi) and abscissa(Ci)
then we set the upper and lower limits
"""
# set working directory and read weights file
# weights were obtained from using Keisan online calculator upto order n=10
path = 'K:/Numerical/chapters/Integration'
os.chdir(path)
gq = pd.read_csv('Gauss10pt.csv') #files with weights and abscissa
xi = gq ['Xi'] #extract absicssa
wi = gq ['Ci'] # extract weights
UL = 1.0 # Upper limit
LL = 0.0 # Lower limit
"""
We call the function newcoord to calculate the second term of the gauss 
quadrature equation.
Then we call our fx function to calculate our first of the gauss quadrature equation
And finally we substitute above terms in quadd quadrature equation and 
calculate the integration and eventually printing our results
"""
zi = newcoord(xi,LL,UL) # get new coords
fzi = fx(zi) # compute function with new coords
result = np.sum(fzi*wi)*(UL-LL)/2 # apply Gauss Quad formula
print(result) #print results






















Q2. WORK DONE
""" 
The work done by the pump is given by
Work Done = Force x Distance
We alraedy hav ethe forec equation and the distance is given by x so our 
equation for work done becomes
Work Done = 2*x**2.5(Force x Distance)
and the limits to integrate this equation are LL=0, UP=10
"""
#import libraries
from scipy.integrate import quadrature
#defining a function to calculate the value of given equation 
def fx(x):
    z = 2*x**2.5 # our given equation
    return z # returning the value backk to function
# integration limits
xo = 0 # lower limit
xn = 10 # upper limit
n = 10 # number of sub-intervals
fxa = fx(xo) # substituting the value of xo in given equaton
fxb = fx(xn) # substituting the value of xn in given equaton
h = (xn-xo)/n # calculating the size of steps
# Finding the sum of fx after substituting the values into our equation
sum = fxa + fxb 
# running a for loop to calculate the value of f(k)
for i in range(1,n):
    k = xo + i*h
    sum = sum + 2 * fx(k) # finding the summation of the term of composite trapezoidal equation
# calculating our final value of composite trapezoidal integration
trap_intigration = 0.5*h*sum   
print('Integration by trapezoidal gives:', trap_intigration) # printing our result

Quad_integration = quadrature(fx, xo, xn)
print('Integration by Gauss Quadrature gives:',Quad_integration)

"""                                           
CONCLUSION:
    1- The Quadrature method given the most accurate approximation when 
    compared to the analytical solution.
    2- When we consider the trapezoidal method, it only gives good 
    approximation when we increase the number of sub-intervals. For low 
    sub-intervals, trapezoidal method doesn't give proper approximation.
Comparison os results:
    analytical solution = 1807.015806
    Quass Quadrature    = 1807.015827
    Trapezoidal method:
        at n = 10 =>    = 1820.207349
        at n = 50 =>    = 1807.542910
        at n = 500 =>   = 1807.021076
"""








Q3.1 Trapezoidal method
'''
Getting assistance from the lecture notes and code provided to calculate the 
intergration of the given equation
'''
#importing libraries
import numpy as np #for numpy library
from scipy.integrate import trapezoid # to use built-in function of trapezoidal method
# Defining the function to be integrated
def fxy(x,y):
    z = (2*x*y + 2*x - x**2 - 2*y**2 +72) # our given equation to be integrated
    return(z) # returns value to function
fxyv = np.vectorize(fxy) # vectorize the function
# Creating an array of points
x = np.arange(0,8.1, 0.1) #to integrate wrt x from 0 to 8 wiht increment of 0.1
y = np.arange(0,6.1,0.1) # to integrate wrt y from 0 to 6 wiht increment of 0.1
xg,yg = np.meshgrid(x,y) # create a grid of points (x,y)
xgf = xg.flatten() # flatten the xg to 1D
ygf = yg.flatten() # flatten the yg to 1D
z = fxyv(xgf,ygf) # compute the values of function at agf and ygf
Nx = xg.shape[1] 
Ny = xg.shape[0]
zz = z.reshape(Ny,Nx) # reshape the flatten data into 2D array
"""
Next we integrate wrt to x
We create 'ny' so that we can run a for loop through each row and then 'F'
to store values after integration
We start a for loop which extracts ith rows from 'zz' and then we integrate 
using in-built trapezoidal function 
"""
# Integrate each column of X
ny = np.arange(0,Ny,1) # loop over each row
F = np.zeros(Ny) # store values corresponding to each row
for i in ny :
    zzy = zz[i,:] # extract ith row
    F[i] = trapezoid(zzy,x) # integrate wrt x
"""
Finally we integrate wrt Y using our built-in function for trapezoidal memthod
amd print our result
"""
# Integrate wrt x
Fxy =trapezoid(F,y)
print(Fxy) # print the results




Q3.2 Romberg Method
"""
Getting assistance from the lecture notes and code provided to calculate the 
intergration of the given equation
"""
#importing libraries
from scipy.integrate import romberg #inbuilt romberg function
from scipy.interpolate import CubicSpline # in-built cubicspline function
import numpy as np
#define a function to calculate integrand
def fxy(x,y): 
   z = (2*x*y + 2*x - x**2 - 2*y**2 +72) # our given equation
   return(z)
""" 
To integrate wrt x we create an array and caculate its length
using its length we initialize an array to store values of integrands
and the run the for loop for every elements in 'Y' and use built-in romberg 
funtion to calculate integrand wrt x
"""
# Integrate with respect to x
# create a sequence of y
Y = np.arange(0,6.1,0.1) # arrange elements from 0-6  with 0.1 imcrement
Ny = len(Y) # length of the array
gxy = np.zeros(Ny) # initialize an array to store integral wrt x values
idx = 0 # initialize the index
for i in Y : # loop through various values of Y
  gxy[idx] = romberg(fxy,a=0,b=8,args=(i,)) # obtain integral value
  idx = idx + 1 # update the index
"""
Now we integrate the with respect to Y
here we are dealing with two variables (x and y) in the equation and to use the 
quadrature method we need to convert it into a equation wiht one varibale, hence
we use cubic splines to do that.
If we do not use the cubicsplines function we won't be able to integrate wrt y
And finally we use built-in function of romberg method to find the integration
and the print our value.
"""
# Integrate with respect to Y
Fxy = CubicSpline(Y,gxy) # Fit a eqn between Y and integrated values wrt x
F = romberg(Fxy,a=0,b=6) # Integrate wrt Y
print(F) # print the result




Q3.3 Trapezoidal + Quadrature
"""
Getting assistance from the lecture notes and code provided to calculate the 
intergration of the given equation
"""
#importing libraries
import numpy as np #for numpy library
from scipy.integrate import trapezoid # to use built-in function of trapezoidal method
from scipy.integrate import quadrature # to use built-in function of Quadrature method
from scipy.interpolate import CubicSpline # in built cubicsplines function 

# Defining the function to be integrated
def fxy(x,y):
    z = (2*x*y + 2*x - x**2 - 2*y**2 +72) # our given equation to be integrated
    return(z) # returns value to function
fxyv = np.vectorize(fxy) # vectorize the function
# Creating an array of points
x = np.arange(0,8.1,0.1) # creates a NDARRAY from 0-8 with intervals of 0.1
y = np.arange(0,6.1,0.1) # creates a NDARRAY from 0-6 with intervals of 0.1
xg,yg = np.meshgrid(x,y) # create a grid of points (x,y)
xgf = xg.flatten() # flatten the xg to 1D
ygf = yg.flatten() # flatten the yg to 1D
z = fxyv(xgf,ygf) # compute the values of function at agf and ygf
Nx = xg.shape[1] 
Ny = xg.shape[0]
zz = z.reshape(Ny,Nx) # reshape the flatten data into 2D array of size Nx x Ny
"""
Next we integrate wrt to x
We create 'ny' so that we can run a for loop through each row and then 'F'
to store values after integration
We start a for loop which extracts ith rows from 'zz' and then we integrate 
using in-built trapezoidal function 
"""
# Integrate each column of X
ny = np.arange(0,Ny,1) # loop over each row
F = np.zeros(Ny) # store values corresponding to each row
for i in ny :
    zzy = zz[i,:] # extract ith row
    F[i] = trapezoid(zzy,x) # integrate wrt x
"""
Now we integrate the with respect to Y
here we are dealing with two variables (x and y) in the equation and to use the 
quadrature method we need to convert it into a equation wiht one varibale, hence
we use cubic splines to do that.
And finally we use built-in function of quadrature method to find the integration
and the print our value.
"""
# Integrate wrt y
Fxy = CubicSpline(y,F) #  fit a equation between Y and integrated values wrt x
f = quadrature(Fxy,0,6) # integrate wrt y
print(f) # print the results


```
